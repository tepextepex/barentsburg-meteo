```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta, date
import os.path
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.dates as mdates
from pandas.tseries.offsets import MonthEnd
import math
```

```python
cm = 1 / 2.54  # inches to cm
wu = 7*cm  # width unit
hu = 7*cm  # height unit

LEAST_SIZE = 7
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

matplotlib.rc('font', size=SMALL_SIZE)          # controls default text sizes
matplotlib.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
matplotlib.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
matplotlib.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
matplotlib.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
matplotlib.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
matplotlib.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
fig_dir = os.path.join(globals()['_dh'][0], "fig")

point_count = (7, 12.5, 14.5, 18, 20, 30, 40, 50, 60, 100)
rms_2016 = (8.1, 4.8, 4.9, 3.9, 2.9, 2.8, 2.1, 1.9, 1.8, 1.2)
rms_2018 = (12.4, 5.0, 6.0, 4.7, 4.6, 3.9, 3.1, 2.5, 2.2, 1.7)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(1*wu, 1*hu)
# plt.xticks(rotation=90)

ax.plot(point_count, rms_2016, label="Snow survey 2016")
ax.plot(point_count, rms_2018, label="Snow survey 2018")

ax.set_xlabel("Number of sampling points")
ax.set_ylabel("RMS between the sample mean\nand the grid-averaged value, cm")

ax.legend()

fig.tight_layout()

fig_paths = ("snow_thickness_uncertainty.png", "snow_thickness_uncertainty.pdf")
for fig_path in fig_paths:
    plt.savefig(os.path.join(fig_dir, fig_path), dpi=300, transparent=False, facecolor="white")
```

#### the same, but normalized by Aldegondabreen area:

```python
area_2018 = 5.54  # sq km

point_density = [float(x) / area_2018 for x in point_count]

fig, ax = plt.subplots()
fig.set_size_inches(1*wu, 1*hu)
# plt.xticks(rotation=90)

ax.plot(point_density, rms_2016, label="Snow survey 2016")
ax.plot(point_density, rms_2018, label="Snow survey 2018")

ax.set_xlabel("Density of sampling points, pt per km$^2$")
ax.set_ylabel("RMS between the sample mean\nand the grid-averaged value, cm")

ax.set_xlim(0, 20)
ax.set_ylim(0, 15)

ax.legend()

plt.minorticks_on()
ax.grid(b=True, which='major', color='grey', linestyle='-')
ax.grid(b=True, which='minor', color='lightgrey', linestyle='--')

fig.tight_layout()

fig_paths = ("snow_thickness_uncertainty_norm.png", "snow_thickness_uncertainty_norm.pdf")
for fig_path in fig_paths:
    plt.savefig(os.path.join(fig_dir, fig_path), dpi=300, transparent=False, facecolor="white", bbox_inches='tight')
```
